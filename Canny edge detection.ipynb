# Import the opencv and numpy library


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# selecting an image through openCV

```python
img = cv2.imread('sudoku.png')
cv2.namedWindow('image1', cv2.WINDOW_NORMAL)   #create a window
cv2.imshow('image1',img)                      #load image to the created window
cv2.waitKey()
cv2.destroyAllWindows()
```

# ploting an image through matplotlib

```python
plt.imshow(img)
```

```python
plt.imshow(img,cmap = 'gray') #, interpolation = 'bicubic')
plt.xticks([]),plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```

# convert the image to gray and apply the canny edge detection s

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap = 'gray') #interpolation = 'bicubic')
plt.xticks([]),plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```

```python
edges = cv2.Canny(gray,10,200,apertureSize = 3)
plt.imshow(edges,cmap = 'gray') #interpolation = 'bicubic')
plt.xticks([]),plt.yticks([]) 
plt.show() # to hide tick values on X and Y axis
#cv2.imshow('image',edges)                      #load image to the created window
#cv2.waitKey()
#cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(gray,10,200,apertureSize = 3)
plt.imshow(edges,cmap = 'gray',interpolation = 'bicubic')
plt.xticks([]),plt.yticks([]) 
plt.show()
```

```python
edges = cv2.Canny(gray,10,200,apertureSize = 3)
cv2.imshow('image',edges)                      #load image to the created window
cv2.waitKey()
cv2.destroyAllWindows()
```
